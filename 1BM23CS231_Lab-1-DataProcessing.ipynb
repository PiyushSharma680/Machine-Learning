### 1. Data Cleaning: Handling Missing Values (Adult Income Dataset)

```python
if not df_adult.empty:
    # Missing values in this dataset are often represented as '?' or other placeholders.
    # First, replace these placeholders with NaN.
    df_adult_cleaned = df_adult.replace('?', np.nan).copy()

    print("Missing values in Adult Income dataset before handling:")
    display(df_adult_cleaned.isnull().sum())

    # Handle numerical missing values with the mean and categorical with the mode
    numerical_cols_adult = df_adult_cleaned.select_dtypes(include=np.number).columns.tolist()
    for col in numerical_cols_adult:
        if df_adult_cleaned[col].isnull().any():
            mean_val = df_adult_cleaned[col].mean()
            df_adult_cleaned[col].fillna(mean_val, inplace=True)
            print(f"Filled numerical missing values in '{col}' with mean: {mean_val:.2f}")

    categorical_cols_adult = df_adult_cleaned.select_dtypes(include='object').columns.tolist()
    for col in categorical_cols_adult:
        if df_adult_cleaned[col].isnull().any():
            # For categorical, it's safer to fill with the mode
            mode_val = df_adult_cleaned[col].mode()[0]
            df_adult_cleaned[col].fillna(mode_val, inplace=True)
            print(f"Filled categorical missing values in '{col}' with mode: {mode_val}")

    print("\nMissing values in Adult Income dataset after handling:")
    display(df_adult_cleaned.isnull().sum())
    print("\nFirst 5 rows after handling missing values:")
    display(df_adult_cleaned.head())
```

### 2. Data Cleaning: Handling Categorical Data (Adult Income Dataset)

```python
if not df_adult_cleaned.empty:
    categorical_cols_adult_encoding = df_adult_cleaned.select_dtypes(include='object').columns.tolist()

    if categorical_cols_adult_encoding:
        print(f"Categorical columns identified for encoding: {categorical_cols_adult_encoding}")
        # Apply one-hot encoding using pd.get_dummies
        df_adult_encoded = pd.get_dummies(df_adult_cleaned, columns=categorical_cols_adult_encoding, drop_first=True)
        print("\nDataset after one-hot encoding categorical features (first 5 rows):")
        display(df_adult_encoded.head())
    else:
        print("No categorical columns found for encoding in Adult Income dataset.")
        df_adult_encoded = df_adult_cleaned.copy()

    print("\nInformation after encoding categorical data:")
    df_adult_encoded.info()
```

### 3. Data Cleaning: Handling Outliers (Adult Income Dataset)

```python
if 'df_adult_encoded' in locals() and not df_adult_encoded.empty:
    # We'll use the IQR (Interquartile Range) method to detect and handle outliers.
    # Numerical columns
    numerical_features_adult = df_adult_encoded.select_dtypes(include=np.number).columns.tolist()

    print("\nDetecting and handling outliers using IQR method for Adult Income dataset...")

    df_adult_outliers_handled = df_adult_encoded.copy()

    for column in numerical_features_adult:
        Q1 = df_adult_outliers_handled[column].quantile(0.25)
        Q3 = df_adult_outliers_handled[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Count outliers
        outliers_count = df_adult_outliers_handled[(df_adult_outliers_handled[column] < lower_bound) | (df_adult_outliers_handled[column] > upper_bound)].shape[0]
        if outliers_count > 0:
            print(f"Column '{column}': {outliers_count} outliers detected (below {lower_bound:.2f} or above {upper_bound:.2f})")
            # Capping outliers: replace values outside the bounds with the bounds themselves
            df_adult_outliers_handled[column] = np.where(df_adult_outliers_handled[column] < lower_bound, lower_bound, df_adult_outliers_handled[column])
            df_adult_outliers_handled[column] = np.where(df_adult_outliers_handled[column] > upper_bound, upper_bound, df_adult_outliers_handled[column])
            print(f"  Outliers in '{column}' have been capped.")
        else:
            print(f"Column '{column}': No significant outliers detected.")

    print("\nDescriptive statistics after outlier handling (Adult Income dataset):")
    display(df_adult_outliers_handled.describe())
```

### 4. Data Transformations: Min-Max Scaler (Normalization) (Adult Income Dataset)

```python
if 'df_adult_outliers_handled' in locals() and not df_adult_outliers_handled.empty:
    # Initialize MinMaxScaler
    minmax_scaler_adult = MinMaxScaler()

    # Apply Min-Max Scaling to numerical features
    numerical_features_adult_scaled = df_adult_outliers_handled.select_dtypes(include=np.number).columns.tolist()

    df_adult_minmax_scaled = df_adult_outliers_handled.copy()
    df_adult_minmax_scaled[numerical_features_adult_scaled] = minmax_scaler_adult.fit_transform(df_adult_minmax_scaled[numerical_features_adult_scaled])

    print("\nAdult Income dataset after Min-Max Scaling (first 5 rows):")
    display(df_adult_minmax_scaled.head())
    print("\nDescriptive statistics after Min-Max Scaling (Adult Income dataset):")
    display(df_adult_minmax_scaled.describe())
```

### 5. Data Transformations: Standard Scaler (Adult Income Dataset)

```python
if 'df_adult_outliers_handled' in locals() and not df_adult_outliers_handled.empty:
    # Initialize StandardScaler
    standard_scaler_adult = StandardScaler()

    # Apply Standard Scaling to numerical features
    # Using the same set of numerical features as for Min-Max Scaler
    numerical_features_adult_scaled = df_adult_outliers_handled.select_dtypes(include=np.number).columns.tolist()

    df_adult_standard_scaled = df_adult_outliers_handled.copy()
    df_adult_standard_scaled[numerical_features_adult_scaled] = standard_scaler_adult.fit_transform(df_adult_standard_scaled[numerical_features_adult_scaled])

    print("\nAdult Income dataset after Standard Scaling (first 5 rows):")
    display(df_adult_standard_scaled.head())
    print("\nDescriptive statistics after Standard Scaling (Adult Income dataset):")
    display(df_adult_standard_scaled.describe())
```

```python
# Check for missing values
print("Missing values in Diabetes dataset before handling:")
display(df_diabetes.isnull().sum())

if not df_diabetes.empty:
    # Example strategy: Fill numerical missing values with the mean and categorical with the mode
    # For this dataset, 'Outcome' is the target, other columns are features.
    # Let's assume numerical features might have missing values. We'll fill with mean.

    # Identify numerical columns (excluding 'Outcome')
    numerical_cols = df_diabetes.select_dtypes(include=np.number).columns.tolist()
    if 'Outcome' in numerical_cols: # Assuming 'Outcome' is numerical (0 or 1)
        numerical_cols.remove('Outcome')

    for col in numerical_cols:
        if df_diabetes[col].isnull().any():
            mean_val = df_diabetes[col].mean()
            df_diabetes[col].fillna(mean_val, inplace=True)
            print(f"Filled missing values in '{col}' with mean: {mean_val:.2f}")

    # For any potential categorical columns, fill with mode (if any exist and have NaNs)
    categorical_cols = df_diabetes.select_dtypes(include='object').columns.tolist()
    for col in categorical_cols:
        if df_diabetes[col].isnull().any():
            mode_val = df_diabetes[col].mode()[0]
            df_diabetes[col].fillna(mode_val, inplace=True)
            print(f"Filled missing values in '{col}' with mode: {mode_val}")

    print("\nMissing values in Diabetes dataset after handling:")
    display(df_diabetes.isnull().sum())
    print("\nFirst 5 rows after handling missing values:")
    display(df_diabetes.head())
```
