```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Create the dataset
data = {
    'a1': [True, True, False, False, False, True, True, True, False, False],
    'a2': ['Hot', 'Hot', 'Hot', 'Cool', 'Cool', 'Cool', 'Hot', 'Hot', 'Cool', 'Cool'],
    'a3': ['High', 'High', 'High', 'Normal', 'Normal', 'High', 'High', 'Normal', 'Normal', 'High'],
    'Classification': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'Yes']
}

data

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert categorical data to numerical data
label_encoders = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Split the dataset into features and target
X = df.drop('Classification', axis=1)
y = df['Classification']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier with entropy as the criterion
clf = DecisionTreeClassifier(criterion='entropy')

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_test, y_pred, target_names=['No', 'Yes']))

# Optionally, visualize the decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['No', 'Yes'])
plt.show()

```

### Regression Tree for Petrol Consumption Prediction

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Load the dataset
petrol_df = pd.read_csv('/content/petrol_consumption.csv')

# Display the first few rows and information about the dataset
print("Petrol Consumption Dataset Head:")
display(petrol_df.head())
print("\nPetrol Consumption Dataset Info:")
petrol_df.info()

# Separate features (X) and target (y)
X_petrol = petrol_df.drop('Petrol_Consumption', axis=1)
y_petrol = petrol_df['Petrol_Consumption']

# Split the dataset into training (80%) and testing (20%) sets
X_train_petrol, X_test_petrol, y_train_petrol, y_test_petrol = train_test_split(X_petrol, y_petrol, test_size=0.2, random_state=42)

# Build the Regression Tree model
regressor_petrol = DecisionTreeRegressor(random_state=42)

# Train the regressor on the training data
regressor_petrol.fit(X_train_petrol, y_train_petrol)

# Make predictions on the test data
y_pred_petrol = regressor_petrol.predict(X_test_petrol)

# Evaluate the model
mae_petrol = mean_absolute_error(y_test_petrol, y_pred_petrol)
mse_petrol = mean_squared_error(y_test_petrol, y_pred_petrol)
rmse_petrol = np.sqrt(mse_petrol)

print(f"\nPetrol Consumption Dataset - Mean Absolute Error: {mae_petrol:.2f}")
print(f"Petrol Consumption Dataset - Mean Squared Error: {mse_petrol:.2f}")
print(f"Petrol Consumption Dataset - Root Mean Squared Error: {rmse_petrol:.2f}")
```

### 1. Decision Tree Classifier for Iris Flower Dataset

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_df = pd.read_csv('/content/iris (1).csv')

# Display the first few rows and information about the dataset
print("Iris Dataset Head:")
display(iris_df.head())
print("\nIris Dataset Info:")
iris_df.info()

# Separate features (X) and target (y)
X_iris = iris_df.drop('species', axis=1)  # Assuming 'species' is the target column
y_iris = iris_df['species']

# Encode the target variable if it's categorical (which 'species' usually is)
le_iris = LabelEncoder()
y_iris_encoded = le_iris.fit_transform(y_iris)

# Split the dataset into training (80%) and testing (20%) sets
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris_encoded, test_size=0.2, random_state=42)

# Build the Decision Tree Classifier
clf_iris = DecisionTreeClassifier(random_state=42)

# Train the classifier on the training data
clf_iris.fit(X_train_iris, y_train_iris)

# Make predictions on the test data
y_pred_iris = clf_iris.predict(X_test_iris)

# Display accuracy score
accuracy_iris = accuracy_score(y_test_iris, y_pred_iris)
print(f"\nIris Dataset - Accuracy Score: {accuracy_iris:.2f}")

# Display confusion matrix
cm_iris = confusion_matrix(y_test_iris, y_pred_iris)
print("\nIris Dataset - Confusion Matrix:")
# Use ConfusionMatrixDisplay for better visualization
display_cm_iris = ConfusionMatrixDisplay(confusion_matrix=cm_iris, display_labels=le_iris.classes_)
display_cm_iris.plot(cmap=plt.cm.Blues)
plt.title('Iris Dataset Confusion Matrix')
plt.show()
```

### 2. Decision Tree Classifier for Drug Dataset

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the Drug dataset
drug_df = pd.read_csv('/content/drug.csv')

# Display the first few rows and information about the dataset
print("Drug Dataset Head:")
display(drug_df.head())
print("\nDrug Dataset Info:")
drug_df.info()

# Identify categorical columns for encoding (assuming object type columns are categorical)
categorical_cols_drug = drug_df.select_dtypes(include=['object']).columns

# Apply Label Encoding to categorical features and the target variable
label_encoders_drug = {}
for col in categorical_cols_drug:
    le = LabelEncoder()
    drug_df[col] = le.fit_transform(drug_df[col])
    label_encoders_drug[col] = le

# Separate features (X) and target (y)
# Assuming the last column is the target, which is common for drug.csv (Drug column)
X_drug = drug_df.iloc[:, :-1]
y_drug = drug_df.iloc[:, -1]

# Split the dataset into training (80%) and testing (20%) sets
X_train_drug, X_test_drug, y_train_drug, y_test_drug = train_test_split(X_drug, y_drug, test_size=0.2, random_state=42)

# Build the Decision Tree Classifier
clf_drug = DecisionTreeClassifier(random_state=42)

# Train the classifier on the training data
clf_drug.fit(X_train_drug, y_train_drug)

# Make predictions on the test data
y_pred_drug = clf_drug.predict(X_test_drug)

# Display accuracy score
accuracy_drug = accuracy_score(y_test_drug, y_pred_drug)
print(f"\nDrug Dataset - Accuracy Score: {accuracy_drug:.2f}")

# Display confusion matrix
cm_drug = confusion_matrix(y_test_drug, y_pred_drug)
print("\nDrug Dataset - Confusion Matrix:")
# If the target was encoded, get the original class names from the LabelEncoder for the target column
target_column_name_drug = drug_df.columns[-1] # Assuming the last column was the target before encoding
if target_column_name_drug in label_encoders_drug:
    display_labels_drug = label_encoders_drug[target_column_name_drug].classes_
else:
    display_labels_drug = None # Fallback if target was not encoded or not found

display_cm_drug = ConfusionMatrixDisplay(confusion_matrix=cm_drug, display_labels=display_labels_drug)
display_cm_drug.plot(cmap=plt.cm.Blues)
plt.title('Drug Dataset Confusion Matrix')
plt.show()
```
